# 分析詞頻分佈

```python
import sqlite3
import matplotlib.pyplot as plt
```

```python
def draw(x, y):
    plt.plot(x, y, label='wf')
    plt.title("word frequency")
    plt.xlabel("rank")
    plt.ylabel("number")
    plt.legend(loc="best")
    plt.show()
```

```python
dbWordFrequency = './word-frequency/word-frequency.db'


conn = sqlite3.connect(dbWordFrequency)
cursor = conn.cursor()

sql = "SELECT * from frequency ORDER BY num DESC"
res = cursor.execute(sql).fetchall()

print(res[:50])
```

```python
sample = 1000
shift = len(res) // sample
x, y = [], []
for i in range(sample):
    x.append(i/sample)
    y.append(res[shift*i][1])
```

```python
draw(x, y)
```

```python
draw(x[:3], y[:3])
```

```python
draw(x[3:150], y[3:150])
print(y[3: 10])
```

```python
draw(x[10:150], y[10:150])
```

```python
draw(x[20:150], y[20:150])
```

```python
draw(x[150:350], y[150:350])
```

```python
draw(x[350:], y[350:])
```
